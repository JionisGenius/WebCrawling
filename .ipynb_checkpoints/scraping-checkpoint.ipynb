```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
driver=webdriver.Chrome(ChromeDriverManager().install())
```

```python
url="https://www.melon.com/chart/index.htm"
```

```python
driver.get(url)
```

```python
from bs4 import BeautifulSoup
```

```python
html=driver.page_source
soup=BeautifulSoup(html, 'html.parser')
```

```python
soup
```

```python
songs=soup.select("table > tbody > tr")
```

```python
len(songs) #1~100
```

```python
rank=0
song_data=[]
for song in songs:
    title=song.select("span > a")[0].text #.text .string
    #print(song.select_one("div.rank02 > a").text)
    singer=song.select("div.rank02 > a")[0].text
    rank=rank+1
    song_data.append(['Melon',rank,title,singer])
#     title=song.select("span > a") ??? #노래제목
#     singer=?? #가수명
```

```python
song_data
```

```python
import pandas as pd
```

```python
melon_df=pd.DataFrame(song_data, columns=['서비스', '순위', '타이틀', '가수'])
```

```python
melon_df.to_excel("melon.xlsx", index=False)
```

```python
html=driver.page_source
soup=BeautifulSoup(html, 'html.parser')
songs=soup.select("table>tbody>tr")
songs
```

```python
# <div class="ellipsis rank02">
# div.ellipsis.rank02
# div.rank02
```

```python
#songs[0].select("a.title.ellipsis")
#songs[0].select("a.title")
#songs[0].select("a.ellipsis") 이건 사용 불가

title=songs[0].select("a.title")[0].text.strip()
```

```python
singer=songs[0].select("a.artist")[0].text.strip()
singer
```

```python
song_data=[]
rank=1
songs=soup.select("table>tbody>tr")
songs
```

```python
for song in songs:
    title=song.select("a.title")[0].text.strip()
    singer=song.select("a.artist")[0].text.strip()
    song_data.append(['Genie', rank, title, singer])
    rank+=1
```

```python
genie_df=pd.DataFrame(song_data, columns=['서비스','순위','타이틀','가수'])
genie_df
genie_df.to_excel("genie.xlsx", index=False)
```
